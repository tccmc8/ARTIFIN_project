```python
num_feat = ['study_hours_per_day','social_media_hours','netflix_hours',
            'attendance_percentage','sleep_hours','exercise_frequency',
            'mental_health_rating','exam_score']

n = len(num_feat)
ncols = 3
nrows = (n + ncols - 1) // ncols

fig, axes = plt.subplots(nrows, ncols, figsize=(18, nrows * 4))
axes = axes.flatten()

for i, col in enumerate(num_feat):
    ax = axes[i]
    data = df[col].dropna()
    sns.histplot(data, kde=True, ax=ax, color='steelblue', bins=30, edgecolor='white')
    ax.axvline(data.mean(),   color='red',    lw=1.8, linestyle='--', label=f'Mean={data.mean():.2f}')
    ax.axvline(data.median(), color='orange', lw=1.8, linestyle='-',  label=f'Median={data.median():.2f}')
    ax.set_title(col, fontsize=11, fontweight='bold')
    ax.set_xlabel('')
    ax.legend(fontsize=7)
    skew = data.skew()
    kurt = data.kurtosis()
    ax.text(0.98, 0.95, f'Skew={skew:.2f}\nKurt={kurt:.2f}',
            transform=ax.transAxes, fontsize=8, ha='right', va='top',
            bbox=dict(facecolor='lightyellow', alpha=0.7))

for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

fig.suptitle('Univariate Distributions — All Numerical Features',
             fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('/tmp/univariate.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
cat_vs_score = ['gender','diet_quality','part_time_job',
                'internet_quality','parental_education_level',
                'extracurricular_participation','attendance_category']

fig, axes = plt.subplots(2, 4, figsize=(22, 10))
axes = axes.flatten()

order_map = {
    'diet_quality'            : ['Poor','Fair','Good','Excellent'],
    'parental_education_level': ['High School','Bachelor','Master','PhD'],
    'internet_quality'        : ['Poor','Average','Good','Excellent'],
    'attendance_category'     : ['Critical','Low','Moderate','High']
}

for i, col in enumerate(cat_vs_score):
    ax = axes[i]
    order = order_map.get(col, None)
    sns.boxplot(data=df, x=col, y='exam_score', ax=ax, palette='Set3',
                order=order)
    # group means
    gm = df.groupby(col)['exam_score'].mean()
    ax.set_title(f'{col}', fontsize=10, fontweight='bold')
    ax.set_xlabel('')
    ax.set_ylabel('Exam Score')
    plt.setp(ax.get_xticklabels(), rotation=25, ha='right', fontsize=8)
    # annotate means
    for tick, label in enumerate(ax.get_xticklabels()):
        val = label.get_text()
        if val in gm.index:
            ax.text(tick, gm[val]+1, f'{gm[val]:.1f}', ha='center',
                    fontsize=8, color='darkblue', fontweight='bold')

axes[-1].set_visible(False)
fig.suptitle('Categorical Features vs Exam Score (Box Plots with Group Means)',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('/tmp/bivariate_categorical.png', dpi=120, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define bins and labels
co2_bins = [0, 11, 30, np.inf]
labels = ["Target", "Average", "Worrisome"]

# Create Carbon Footprint Category
df["CO2_Category"] = pd.cut(
    df["Carbon_Footprint_kgCO2"],
    bins=co2_bins,
    labels=labels,
    right=False
)

# Create figure
plt.figure(figsize=(15, 12))

# =======================
# 1️⃣ KDE Plot
# =======================
plt.subplot(2, 1, 1)

sns.kdeplot(
    x="Carbon_Footprint_kgCO2",
    data=df,
    fill=True,
    color="seagreen",
    linewidth=2,
    alpha=0.7
)

plt.title("Carbon Footprint Distribution (KDE Plot)", fontsize=16, fontweight="bold")
plt.xlabel("Carbon Footprint (kg CO2)", fontsize=14)
plt.ylabel("Density", fontsize=14)

mean_co2 = df["Carbon_Footprint_kgCO2"].mean()

xticks = list(range(0, int(df["Carbon_Footprint_kgCO2"].max()) + 10, 10)) + [mean_co2]
plt.xticks(xticks)

plt.axvline(x=mean_co2, color="red", linestyle="--", label="Mean", alpha=0.6)

plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()

plt.tight_layout()

# =======================
# 2️⃣ Count Plot
# =======================
plt.subplot(2, 1, 2)

ax = sns.countplot(
    x="CO2_Category",
    data=df,
    palette="coolwarm",
    edgecolor="black",
    linewidth=1.5
)

plt.title("Carbon Footprint Category Distribution", fontsize=16, fontweight="bold")
plt.xlabel("CO2 Category", fontsize=14)
plt.ylabel("Number of Individuals", fontsize=14)

# Add value labels on bars
for p in ax.patches:
    ax.annotate(
        f"{p.get_height()}",
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha="center",
        va="center",
        fontsize=12,
        color="black",
        xytext=(0, 9),
        textcoords="offset points"
    )

plt.xticks(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Drop missing values
co2_data = df.dropna(subset=["CO2_Category", "Carbon_Footprint_kgCO2"])

# Group by CO2 category
co2_groups = co2_data.groupby("CO2_Category")["Carbon_Footprint_kgCO2"]

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

co2_data.boxplot(
    column="Carbon_Footprint_kgCO2",
    by="CO2_Category",
    ax=ax,
    grid=False,
    patch_artist=True,
    boxprops=dict(facecolor="lightgreen", color="green"),
    medianprops=dict(color="red"),
    whiskerprops=dict(color="green"),
    capprops=dict(color="green"),
    flierprops=dict(markerfacecolor='gray',
                    marker='o',
                    markersize=5,
                    linestyle='none')
)

# Optional: Calculate summary statistics (like your example)
for i, (category, group) in enumerate(co2_groups):
    mean = group.mean()
    std = group.std()
    min_val = group.min()
    max_val = group.max()
    print(f"{category}:")
    print(f"  Mean: {mean:.2f}")
    print(f"  Std: {std:.2f}")
    print(f"  Min: {min_val:.2f}")
    print(f"  Max: {max_val:.2f}")
    print("-" * 30)

# Labels
ax.set_title("Carbon Footprint by CO2 Category")
ax.set_xlabel("CO2 Category")
ax.set_ylabel("Carbon Footprint (kg CO2)")

plt.suptitle("")  # Remove automatic subtitle
plt.show()
```
