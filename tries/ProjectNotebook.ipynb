```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Copy dataset and drop citizen_id
df_focus = df.drop(columns=["citizen_id"], errors="ignore")

# Convert CO2 category to numeric
df_focus["CO2_Category_Code"] = df_focus["CO2_Category"].cat.codes

# Encode categorical variables
for col in df_focus.select_dtypes(include=["object", "category"]).columns:
    df_focus[col] = df_focus[col].astype("category").cat.codes

# Correlation matrix
corr_matrix = df_focus.corr()

# Extract only correlations with CO2 category
co2_corr = corr_matrix[["CO2_Category_Code"]].sort_values(
    by="CO2_Category_Code", ascending=False
)

plt.figure(figsize=(8, 12))

sns.heatmap(
    co2_corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation with CO2 Category", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# Drop citizen_id
df_group = df.drop(columns=["citizen_id"], errors="ignore")

# Encode categorical columns except CO2_Category
df_group_encoded = df_group.copy()

for col in df_group_encoded.select_dtypes(include=["object"]).columns:
    if col != "CO2_Category":
        df_group_encoded[col] = df_group_encoded[col].astype("category").cat.codes

# Compute group means
group_means = df_group_encoded.groupby("CO2_Category").mean()

plt.figure(figsize=(18, 8))

sns.heatmap(
    group_means,
    annot=True,
    cmap="YlGnBu",
    fmt=".1f",
    linewidths=0.5
)

plt.title("Average Feature Values per CO2 Category", fontsize=16, fontweight="bold")
plt.xlabel("Features")
plt.ylabel("CO2 Category")

plt.tight_layout()
plt.show()
```

```python
# Prepare clustering dataset
df_cluster = df.drop(columns=["citizen_id"], errors="ignore")

# Encode categorical variables
for col in df_cluster.select_dtypes(include=["object", "category"]).columns:
    df_cluster[col] = df_cluster[col].astype("category").cat.codes

# Standardize data (important for clustering)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_cluster)

# Convert back to DataFrame
df_scaled = pd.DataFrame(scaled_data, columns=df_cluster.columns)

# Create clustered heatmap
sns.clustermap(
    df_scaled,
    cmap="coolwarm",
    figsize=(14, 12),
    metric="euclidean",
    method="ward"
)

plt.title("Clustered Heatmap of Citizen Behavior Patterns")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create age groups
labels = ['Young', 'Middle-aged', 'Elderly']
bins = [18, 30, 50, 70]

df['Age_Group'] = pd.cut(
    df['Age'],
    bins=bins,
    labels=labels,
    right=False
)
```

```python
# Create cross-tab
age_co2 = pd.crosstab(
    df['Age_Group'],
    df['CO2_Category'],
    normalize='index'
) * 100

plt.figure(figsize=(8,6))

sns.heatmap(
    age_co2,
    annot=True,
    cmap="YlOrRd",
    fmt=".1f"
)

plt.title("CO2 Category Distribution by Age Group (%)", fontsize=14, fontweight='bold')
plt.ylabel("Age Group")
plt.xlabel("CO2 Category")

plt.tight_layout()
plt.show()
```

```python
# Drop missing values
co2_data = df.dropna(subset=["CO2_Category", "Carbon_Footprint_kgCO2"])

# Group by CO2 category
co2_groups = co2_data.groupby("CO2_Category")["Carbon_Footprint_kgCO2"]

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

co2_data.boxplot(
    column="Carbon_Footprint_kgCO2",
    by="CO2_Category",
    ax=ax,
    grid=False,
    patch_artist=True,
    boxprops=dict(facecolor="lightgreen", color="green"),
    medianprops=dict(color="red"),
    whiskerprops=dict(color="green"),
    capprops=dict(color="green"),
    flierprops=dict(markerfacecolor='gray',
                    marker='o',
                    markersize=5,
                    linestyle='none')
)

# Optional: Calculate summary statistics (like your example)
for i, (category, group) in enumerate(co2_groups):
    mean = group.mean()
    std = group.std()
    min_val = group.min()
    max_val = group.max()
    print(f"{category}:")
    print(f"  Mean: {mean:.2f}")
    print(f"  Std: {std:.2f}")
    print(f"  Min: {min_val:.2f}")
    print(f"  Max: {max_val:.2f}")
    print("-" * 30)

# Labels
ax.set_title("Carbon Footprint by CO2 Category")
ax.set_xlabel("CO2 Category")
ax.set_ylabel("Carbon Footprint (kg CO2)")

plt.suptitle("")  # Remove automatic subtitle
plt.show()
```
